```python
import pandas as pd
from geopy.geocoders import Nominatim
import ast
```

```python
PATH = '../../data/'
```

```python
coffee_shops = pd.read_csv(PATH + 'raw/' + 'sg_coffee_shops.csv')
coffee_shops.head()
```

```python
coffee_shops.info()
```

### Clean Price

```python
# expand out categories to get the type of shop
coffee_shops['categories'] = coffee_shops['categories'].apply(lambda x: [d['alias'] for d in ast.literal_eval(x)])
coffee_shops.head()
```

```python
coffee_shops['joined_cat'] = coffee_shops['categories'].apply(lambda x: ' '.join(x))
coffee_shops['joined_cat'].value_counts()
```

```python
# store mode of each unique shop category and replace NaN price with them
cat_mode_price = {}
for k,v in coffee_shops.groupby('joined_cat')['price'].agg(pd.Series.mode).items():
    if isinstance(v, str):
        cat_mode_price[k] = v
    else:
        cat_mode_price[k] = None

coffee_shops['price'] = coffee_shops.apply(lambda row: cat_mode_price[row['joined_cat']] if not isinstance(row['price'],str) else row['price'], axis=1)
```

```python
coffee_shops[['price']].info()
```

```python
# replace remaining NaN price as follows:
# loop through its categories and take the mode of that category if it exist in dict
# else replace it with mode of cafes, which happens the most in the data 

def replace_price_with_first_cat(x):
    if not isinstance(x['price'], str):
        cats = x['categories']
        for c in cats:
            if cat_mode_price.get(c):
                return cat_mode_price[c]
        return cat_mode_price['cafes']
    else:
        return x['price']
```

```python
coffee_shops['price'] = coffee_shops.apply(lambda row: replace_price_with_first_cat(row), axis=1)
```

```python
coffee_shops[['price']].info()
```

```python
coffee_shops['price'].value_counts()
```

### Clean location and coordinates

```python
tmp = coffee_shops['coordinates'].apply(lambda x: ast.literal_eval(x))
tmp.tolist()
```

```python
coffee_shops['address'] = coffee_shops['location'].apply(lambda x: ', '.join(ast.literal_eval(x).get('display_address')))
coffee_shops['coordinates'] = coffee_shops['coordinates'].apply(lambda x: ast.literal_eval(x))
coffee_shops = pd.concat([coffee_shops, pd.DataFrame(coffee_shops['coordinates'].tolist())], axis=1)
```

```python
coffee_shops.head()
```

```python
cols = [
    'id', 'alias', 'name', 'image_url', 'is_closed', 'url', 'review_count',
    'categories', 'rating', 'price', 'address', 'display_phone', 
    'latitude', 'longitude', 'distance',
]
coffee_shops = coffee_shops[cols]
coffee_shops.head()
```

```python
coffee_shops.drop_duplicates(subset='alias', inplace=True)
coffee_shops.info()
```

```python
coffee_shops.to_csv(PATH + 'processed/' + 'sg_coffee_shops_final.csv', index=False)
```

### Converting to One-hot for Context Features

```python
cs_context = coffee_shops.copy()

price_map = {'$':'1', '$$':'2', '$$$':'3', '$$$$':'4'}
cs_context['price'] = cs_context['price'].map(price_map)
cs_context = pd.get_dummies(cs_context, prefix='price', columns=['price'])
```

```python
cat_counts = {}
for l in cs_context.categories:
    for c in l:
        cat_counts[c] = cat_counts.get(c, 0) + 1

cats = list({k:v for k,v in cat_counts.items() if v > 20}.keys())
cats.remove('coffee')
cats.append('coffeeroasteries')
cats.append('themedcafes')

cat_dummy = pd.get_dummies(cs_context['categories'].apply(pd.Series).stack()).sum(level=0)
cat_dummy = cat_dummy[cat_dummy.columns.intersection(cats)]
cs_context = pd.concat([cs_context, cat_dummy], axis=1)
```

```python
loc_bin = pd.read_csv('../../data/processed/coffee_binned.csv')[['alias', 'region']]
loc_bin['region'] = loc_bin['region'].apply(lambda x: x.replace(' ',''))

cs_context = cs_context.merge(loc_bin, on='alias', how='left')
cs_context.drop_duplicates(subset='alias', inplace=True)

cs_context = pd.get_dummies(cs_context, prefix='', prefix_sep='', columns=['region'])
```

```python
cs_context.drop([
    'id', 'name', 'image_url', 'is_closed', 'url', 'review_count', 
    'categories', 'rating', 'address', 'display_phone', 'latitude', 
    'longitude', 'distance'],
    axis=1, inplace=True)

cs_context.head()
```

```python
cs_context.to_csv(PATH + 'processed/' + 'coffee_shops_context.csv', index=False)
```
